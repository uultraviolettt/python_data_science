```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tables = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C",flavor='html5lib')
```

```python
print(tables)
```

Удалите из таблицы данные по всей стране

```python
my_table = tables[10][:-1]
display(my_table)
```

Вывести первые строки таблицы с помощью метода head

```python
my_table.head()
```

Определите количество строк и столбцов в датафрейме.

```python
my_table.shape
```

Замените в таблице значения "—" на значения NaN

```python
my_table = my_table.replace('—', np.nan)
display(my_table)
```

Определите типы всех столбцов с помощью

```python
my_table.dtypes
```

Замените типы не числовых колонок на числовые. 

```python
my_table[["2014", "2019"]] = my_table[["2014", "2019"]].apply(pd.to_numeric)
my_table.dtypes
```

Посчитайте, какая доля пропусков содержится в каждой колонке (используйте методы isnull и sum)

```python
my_table.isnull().sum()
```

Замените отсутствующие данные в столбцах средними значениями по этим столбцам

```python
my_table = my_table.fillna({"1950": my_table["1950"].mean(), "1960": my_table["1960"].mean(), "1970": my_table["1950"].mean(), "2014": my_table["2014"].mean(),"2019": my_table["2019"].mean()})
display(my_table)
```

Получите список регионов, где уровень рождаемости в 2019 году был выше среднего по Украине

```python
new_tab = pd.concat([my_table["Регіон"], my_table["2019"]], axis = 1)

new_tab = new_tab[new_tab["2019"] > new_tab["2019"].mean()]
res_list = []
for r in new_tab['Регіон']:
  res_list.append(r)
display(res_list)
```

В каком регионе была самая высокая рождаемость в 2014 году?

```python
new_tab = pd.concat([my_table["Регіон"], my_table["2014"]], axis = 1)
display(new_tab[new_tab["2014"] >=  new_tab['2014'].max()])
```

Постройте столбцовую диаграмму рождаемости по регионам в 2019 году

```python
Region = my_table["Регіон"]
birth_rate = my_table["2019"]

plt.bar(Region, birth_rate)
plt.xticks(rotation=90)
plt.xlabel('Region', fontsize=14)
plt.ylabel('birth_rate', fontsize=14)
plt.title('2019', fontsize=14)
plt.show()
```
